```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Read the image
image = cv2.imread('image1.png')

# Show the original image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Original Image')
plt.show()
```

```python
# Convert to greyscale
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# Show the greyscale image
plt.imshow(grey_image, cmap='gray')
plt.axis('off')
plt.title('Greyscale Image')
plt.show()
```

```python
# Calculate the average greyscale value of the whole image
average_grey = np.mean(grey_image)

# Apply thresholding based on the average greyscale value
_, threshold_image = cv2.threshold(grey_image, average_grey, 255, cv2.THRESH_BINARY)


# Display the thresholded image
plt.imshow(threshold_image, cmap='gray')
plt.axis('off')
plt.title('Resultant Image')
plt.show()


```

```python
_, labels = cv2.connectedComponents(threshold_image)

# Exclude the background label
num_areas = len(np.unique(labels)) - 1

# Print the number of areas
print("Number of areas in the image:", num_areas)
```
