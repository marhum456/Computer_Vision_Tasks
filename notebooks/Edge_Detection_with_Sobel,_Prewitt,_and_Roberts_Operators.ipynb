```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_sobel_operator(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
    gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
    gradient_direction = np.arctan2(sobel_y, sobel_x)
    return gradient_magnitude, gradient_direction

image_path = 'image1.png'
image = cv2.imread(image_path)
sobel_magnitude, sobel_direction = apply_sobel_operator(image)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(sobel_magnitude, cmap='gray')
plt.title('Sobel Magnitude')

plt.subplot(1, 2, 2)
plt.imshow(sobel_direction, cmap='jet')
plt.title('Sobel Direction')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_prewitt_operator(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    prewitt_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
    prewitt_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
    prewitt_x_result = cv2.filter2D(gray_image, cv2.CV_64F, prewitt_x)
    prewitt_y_result = cv2.filter2D(gray_image, cv2.CV_64F, prewitt_y)
    gradient_magnitude = np.sqrt(prewitt_x_result**2 + prewitt_y_result**2)
    gradient_direction = np.arctan2(prewitt_y_result, prewitt_x_result)
    return gradient_magnitude, gradient_direction

image_path = 'image1.png'
image = cv2.imread(image_path)
prewitt_magnitude, prewitt_direction = apply_prewitt_operator(image)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(prewitt_magnitude, cmap='gray')
plt.title('Prewitt Magnitude')

plt.subplot(1, 2, 2)
plt.imshow(prewitt_direction, cmap='jet')
plt.title('Prewitt Direction')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_roberts_cross_operator(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    roberts_x = np.array([[1, 0], [0, -1]])
    roberts_y = np.array([[0, 1], [-1, 0]])
    roberts_x_result = cv2.filter2D(gray_image, cv2.CV_64F, roberts_x)
    roberts_y_result = cv2.filter2D(gray_image, cv2.CV_64F, roberts_y)
    gradient_magnitude = np.sqrt(roberts_x_result**2 + roberts_y_result**2)
    gradient_direction = np.arctan2(roberts_y_result, roberts_x_result)
    return gradient_magnitude, gradient_direction

image_path = 'image1.png'
image = cv2.imread(image_path)
roberts_magnitude, roberts_direction = apply_roberts_cross_operator(image)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(roberts_magnitude, cmap='gray')
plt.title('Roberts Magnitude')

plt.subplot(1, 2, 2)
plt.imshow(roberts_direction, cmap='jet')
plt.title('Roberts Direction')

plt.tight_layout()
plt.show()
```
